```python
import pandas as pd
import networkx as nx
import numpy as np
g = nx.Graph()
In_class_network = []
In_class_network_for_use = []
with open("In-class_network.txt", 'r') as f:
    for line in f:
        everyline = line.replace('\t', 'abc').split('abc')
        everyline = np.array(everyline)
        In_class_network += [everyline]
del In_class_network[0]
In_class_network_new = np.array(In_class_network, dtype = object)
In_class_network_for_use = In_class_network_new[:, 0:2]
In_class_network_for_use[In_class_network_for_use == " "] = "0" 
for line in In_class_network_for_use:
    if line[1] == "0":
        g.add_node(line[0])
    else:
        temp_list = line[1].split(',')
        for templine in temp_list:
            g.add_edge(line[0],templine)
    temp_list.clear()
k = g.copy()
count = 0
pathlist = []
component_size = []
k.remove_nodes_from(list(nx.isolates(k)))
connected_components = nx.connected_component_subgraphs(k)
total_list = []
for component in connected_components:
    if nx.number_of_nodes(component) >= 3:
        #print(list(component))
        #print(nx.number_of_nodes(component))
        #print("------------------")
        for items1 in component:
            #print(items1)
            #print(list(nx.all_neighbors(component,items1)))
            #print("--------")
            for items2 in nx.all_neighbors(component,items1):
                #print(items2)
                #print(list(nx.all_neighbors(component,items2)))
                #print("----")
                for items3 in nx.all_neighbors(component,items2):
                    if items1 in list(nx.all_neighbors(component,items3)):
                        #print(items1,items2,items3)
                        three_list = []
                        three_list.append(int(items1))
                        three_list.append(int(items2))
                        three_list.append(int(items3))
                        three_list.sort()
                        total_list.append(three_list)
total_list = np.array(total_list)    
total_new_list = np.unique(total_list, axis=0)
triangle = total_new_list.shape[0]
print(triangle)
    
```
