```python
import networkx as nx
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_table("In-class_network2.txt")
```

```python
G = nx.Graph()
```

```python
for node , friends in df[["ID","IDs-of-acquaintances"]].values:
    if " " in friends: 
        G.add_node(node)
        continue
    friends = list(map(int,friends.split(",")))
    G.add_edges_from(list(zip([node]*len(friends),friends)))
```

```python
sns.countplot(list(dict(G.degree).values()))
```

```python
plt.subplots(figsize=(20,15))
x,y = zip(*dict(G.degree).items())
sns.barplot(x,y)
```
